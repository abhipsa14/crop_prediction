```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler=StandardScaler()
x_transformed=std_scaler.fit_transform(x)
```

```python
sns.kdeplot(pd.DataFrame(x_transformed)[3])
#scaled data-->outliers hid hojayenge
#k-means algo-->to give accuracy of model
```

```python
## training model
from sklearn.cluster import KMeans
```

```python
kmeans_obj=KMeans() 
###by default 8 groups mei divide karti h data ko
##it is used for making clusters,i.e, groups which then decides which group is best suited according to the conditions provided.
```

```python
labels=kmeans_obj.fit_predict(x_transformed) #algorithm is trained and prediction is also done
```

```python
labels
```

```python
set(labels) #groups mei divide hogaye hain
```

```python
labels.shape
```

```python
df['cluster_no']=labels
```

```python
df.head(29)
```

```python
df[['label','cluster_no']].to_csv('./models/filter_crops.csv',index=False)
#data to save kar skte hain aur aage jake use kar skte hain
```
